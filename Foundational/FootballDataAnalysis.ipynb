```python
salary = fd[["short_name", "wage_eur"]]
salary.head(20) #shows the first 20 records of players with their income in Euros
```

```python
#sorting the players according to their wage/ income
player_salary = salary.sort_values(by=["wage_eur"], ascending=False)
player_salary.head(10) #it displays the top 10 highest paid players
```

```python
plt.figure(figsize=(20,10))
plt.bar(list(player_salary['short_name'])[0:10],list(player_salary['wage_eur'])[0:10],color="purple")
plt.show() #just the graphical representation of the top 10 highest paid players
```

```python
#exactly the same thing just changing the color of individual columns
plt.figure(figsize=(20,10))
plt.bar(list(player_salary['short_name'])[0:10],list(player_salary['wage_eur'])[0:10],color=["purple", "indigo", "navy", "blue","green","yellow","pink", "violet", "brown", "darkred"])
#["purple", "indigo","blue","green","olive","yellow", "orange", "brown", "red", "marron"]
plt.show() #just the graphical representation of the top 10 highest paid players
```
